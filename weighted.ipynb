```python
import pandas as pd
import numpy as np
```

```python
df_credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
df_movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df_credits.head()
```

```python
df_movies.head()
```

```python
print("Credict:", df_credits.shape)
print("Movies:", df_movies.shape)
```

Rename the IDs and merge the Dataset

```python
df_credits_renamed = df_credits.rename(index=str, columns={"movie_id": "id"})
df_movies_merged = df_movies.merge(df_credits_renamed, on="id")
df_movies_merged.head()
```

Remove Redundant Features

```python
df_movies_cleaned = df_movies_merged.drop(columns=['homepage', 'title_x', 'title_y', 'production_companies', 'status'])
df_movies_cleaned.head()
```

```python
df_movies_cleaned.info()
```

```python
df_movies_cleaned.isnull().sum()
```

- Using Weighted average for each movie's Average Rating


- Source: http://trailerpark.weebly.com/imdb-rating.html?source=post_page

Calculate all the components based on the above formula

```python
R = df_movies_cleaned["vote_average"]
v = df_movies_cleaned["vote_count"]
m = df_movies_cleaned["vote_count"].quantile(0.70)
C = df_movies_cleaned["vote_average"].mean()
```

```python
df_movies_cleaned["weighted_average"] = (R*v+C*m)/(v+m)
df_movies_cleaned.head()
```

```python
df_movies_sorted_ranking = df_movies_cleaned.sort_values("weighted_average", ascending=False)
df_movies_sorted_ranking[["original_title", "vote_average", "vote_count", "weighted_average", "popularity"]].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Select top 10
top10 = (df_movies_sorted_ranking.sort_values("weighted_average", ascending=False).head(10))
colors = sns.color_palette("tab10", n_colors=10)
plt.figure(figsize=(12, 6))
ax = sns.barplot(
    data=top10,
    x="weighted_average",
    y="original_title",
    palette=colors
)
plt.xlim(4, 10)
plt.title("Top 10 Movies by Weighted Average Rating", weight="bold", fontsize=20)
plt.xlabel("Weighted Average Score", fontsize=15)
plt.ylabel("Movie Title", fontsize=15)
plt.tight_layout()
plt.savefig("best_weighted_average_movies.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df_movies_popularity = df_movies_sorted_ranking.sort_values("popularity", ascending=False)
plt.figure(figsize=(12, 6))
# Select top 10
top10 = (df_movies_popularity.head(10))
ax1 = sns.barplot(
    data=top10,
    x="popularity",
    y="original_title",
    palette=colors
)

# plt.xlim(20, 250)
plt.title("10 Top popular Movies", weight="bold")
plt.xlabel("Popularity Score", weight="bold")
plt.ylabel("Original Title", weight="bold")
plt.savefig("most_popular_movies.png")
```
